SIMULACION MICROGOTA EVAPORACION CONSIDERANDO UN FLUJO DE AIRE
======================

Ahora resolveremos considerando parte de la gota se evapora y un flujo del aire $\vec{v_f}= v_f \hat{y}$:

$$m\frac{dv}{dt} + v\frac{dm}{dt}= E + F_d - mg = \frac{4\pi}{3} r_d^3 \rho_a g - 6\pi \eta r_d (v - v_f) - \frac{4\pi}{3} r_d^3 \rho_d g $$ 

luego

$$ \frac{dv}{dt}= \left( \frac{\rho_a}{\rho_d} - 1 \right) g - \frac{9 \eta}{2  r^2(t) \rho_d}  (v - v_f)  - \frac{v}{m}\frac{dm}{dt}  $$

$$ \frac{dv}{dt}= \left( \frac{\rho_a}{\rho_d} - 1 \right) g - \frac{9 \eta}{2  r_d^2 e^{-2 \alpha t} \rho_d}  (v - v_f)  - \frac{-3\alpha r_d^3 e^{-3\alpha t}}{ r_d^3 e^{-3\alpha t} } v $$

$$ \frac{dv}{dt}= \left( \frac{\rho_a}{\rho_d} - 1 \right) g - \frac{9 \eta}{2  r_d^2 e^{-2 \alpha t} \rho_d}  (v - v_f)  + 3\alpha v $$


con

$$r(t) = r_d e^{-\alpha t} $$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

pi=np.pi

ro_a= 1.225 # kg/m3
ro_d= 1000 # kg/m3
eta= 1.849e-5 # kg/m.s 
r_d= 5e-6 # m
g= 9.81 # m/s2

alpha= 10
```

```python
ro_v= 0.025 
D= 0.25e-4
r_0= 5e-6
velocidad_flujo= 0.15

y0=[0,0]
```

```python
def gota_con_arrastre(y, t, r_0= 5e-6, alpha1= 10, velocidad_flujo= 0.15):
    
    dy=[0,0]
    dy[0]= y[1] 
    dy[1]= (((ro_a /ro_d) - 1) * g ) - (9 *eta / (2 * ( r_0**2 * np.exp(-2*t*alpha1) * ro_d))) * (y[1] - velocidad_flujo) + 3*alpha1 * (y[1] - velocidad_flujo)
    return dy
```

```python
t_0= 0
t_max= 20
dt= 0.001
t= np.arange(t_0, t_max+dt, dt)

plt.figure()
plt.rcParams['font.size']=12

solucion_con_arrastre= odeint(gota_con_arrastre, y0, t, args=(50e-6,0.25/2,0.15))
plt.plot(t,solucion_con_arrastre[:,0],'r', label='v$_{flujo}$= 0.15 m/s ')
plt.plot(t,0*np.ones(len(t)),':',color='k')

ax=plt.gca()
#ax.set_title('Simulacion gota Z(t); r_0= 50 \N{greek small letter mu}m; \N{greek small letter alpha}= 0.25  ')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Posicion (m)')
ax.legend()
plt.axis()
plt.show()
```

```python
plt.plot(t,solucion_con_arrastre[:,1]*1e2,'r', label='v$_{flujo}$= 0.15 m/s ')
plt.plot(t,0*np.ones(len(t)),':',color='k')

ax=plt.gca()
#ax.set_title('Simulacion gota Vz(t); r_0= 50 \N{greek small letter mu}m; \N{greek small letter alpha}= 0.25  ')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Velocidad (cm/s)')
ax.legend()
plt.axis()
plt.show()
```
